## **Calculus of Variations Review**
---
### We have a functional, $f[y(x),y'(x),x]$, and we wnat to extremize it under some constraints. Typically, we want to find $y(x)$ that minimizes the action integral:
$$
S = \int_a^b f[y(x),y'(x),x] dx
$$
### We begin by defining some variant/perturbed function: $R(x) = y(x) + \alpha \epsilon(x)$ which represnets some function thats away from the true minimzing function for the action integral. We are given the endpoints, thus all function, regardless of inaccuracy, must at least pass through the points, meaning a the endpoints, $\epsilon(x_1)=\epsilon(x_2) = 0$ so that $R(x_i) = y(x_i)$. 
### If the parameter $\alpha=0$ then $R(x) = y(x)$ which is our true solution. By introducing this $\alpha$ parameter, the action now becomes a function of $\alpha$: $S:= S(\alpha)$:
$$
S(\alpha) = \int f[R(x) ,R'(x),x] dx
$$

### at $\alpha = 0 $ the action integral is minimized (more generally extremized) which implies $\frac{dS}{d\alpha} = 0, \alpha =0$. So the integral expression becomes:
$$
\begin{align}
\frac{dS}{d\alpha}|_{\alpha =0}  &= \frac{d}{d\alpha} \left(\int_a^b f[R(x) ,R'(x),x] dx \right) |_{\alpha =0} = 0\nonumber\\
&= \int_a^b  \frac{d}{d\alpha} \left( f[R(x) ,R'(x),x] \right)|_{\alpha =0}dx =0 \\
&\text{only 2 of the variables of the functional $f$ depend on alpha so the total must become a partial  }\nonumber\\
&= \int_a^b  \frac{\partial}{\partial\alpha} \left( f[R(x) ,R'(x),x] \right)|_{\alpha =0}dx = 0\\
\end{align}
$$

### Using our knowledge of the multivariable chain rule with $f = f(x_1, ... , x_n)$:
$$
\frac{\partial f}{\partial \alpha } =\sum_i  \frac{\partial f}{\partial x_i } \frac{\partial x_i }{\partial \alpha } 
$$

### we expand the integrand as follows:
$$
\begin{align}
\frac{\partial}{\partial\alpha} \left( f[R(x) ,R'(x),x] \right)|_{\alpha =0} &= \left( \frac{\partial f}{\partial R}\right) \frac{\partial R}{\partial \alpha}|_{\alpha = 0} + 
\left( \frac{\partial f}{\partial R'}\right) \frac{\partial R'}{\partial \alpha}|_{\alpha = 0} + \left( \frac{\partial f}{\partial x}\right) \frac{\partial x}{\partial \alpha}|_{\alpha = 0}\\
\end{align}
$$

### we make note of the following substitutions:
- ### $\frac{\partial x}{\partial \alpha} = 0$
- ### $\frac{\partial R}{\partial \alpha} = \frac{\partial}{\partial \alpha} \left(  y(x) + \alpha \epsilon(x) \right) = \epsilon(x)$
- ### $\frac{\partial R'}{\partial \alpha} = \frac{\partial}{\partial \alpha} \left(  y'(x) + \alpha \epsilon'(x) \right) = \epsilon'(x)$

$$
\begin{align}
&= \left( \frac{\partial f}{\partial R}\right) \frac{\partial R}{\partial \alpha}|_{\alpha = 0} 
+ \left( \frac{\partial f}{\partial R'}\right) \frac{\partial R'}{\partial \alpha}|_{\alpha = 0} 
+ \left( \frac{\partial f}{\partial x}\right) \frac{\partial x}{\partial \alpha}|_{\alpha = 0}\\

&= \left(\epsilon(x) \frac{\partial f}{\partial R}\right) |_{\alpha = 0} 
+ \left( \epsilon'(x)\frac{\partial f}{\partial R'}\right) |_{\alpha = 0} \\
\end{align}
$$

### recall that the action integral is minimzed at $\alpha = 0$, so $R(x) \rightarrow R(x) = y(x) + 0\times \epsilon(x)$ and the same is for $R'(x)$:
$$
\begin{align}
\frac{\partial}{\partial\alpha} \left( f[R(x) ,R'(x),x] \right)|_{\alpha =0}&= \left(\epsilon(x) \frac{\partial f}{\partial y}\right) 
+ \left( \epsilon'(x)\frac{\partial f}{\partial y'}\right) \\
\end{align}
$$

### one last thing to note before integration, the second term: $ \left( \epsilon'(x)\frac{\partial f}{\partial y'}\right)$ can be seen from a result of a product rule:
$$
\begin{align}
\frac{d }{dx}\left( \epsilon(x)\frac{\partial f}{\partial y'} \right) &= \epsilon'(x)\frac{\partial f}{\partial y'} + \epsilon(x) \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right)\\
\epsilon'(x)\frac{\partial f}{\partial y'} + \epsilon(x) &= \frac{d }{dx}\left( \epsilon(x)\frac{\partial f}{\partial y'} \right) -\epsilon(x) \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right)
\end{align}
$$

### So the partial w.r.t. $\alpha$ of the functional becomes:

$$
\frac{\partial}{\partial\alpha} \left( f[R(x) ,R'(x),x] \right)|_{\alpha =0} = \left(\epsilon(x) \frac{\partial f}{\partial y}\right) +\frac{d }{dx}\left( \epsilon(x)\frac{\partial f}{\partial y'} \right) -\epsilon(x) \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right)
$$
### now we can substitute this into the integral for easy evaluation:

$$
\begin{align}
\frac{dS}{d\alpha}|_{\alpha =0}  &=  \int_a^b  \frac{\partial}{\partial\alpha} \left( f[R(x) ,R'(x),x] \right)|_{\alpha =0}dx  = 0\nonumber\\
&= \int_a^b\left( \left(\epsilon(x) \frac{\partial f}{\partial y}\right) +\frac{d }{dx}\left( \epsilon(x)\frac{\partial f}{\partial y'} \right) -\epsilon(x) \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right) \right) dx =0\nonumber\\
&= \int_a^b \epsilon(x)\left(  \frac{\partial f}{\partial y} - \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right) \right) dx + \int_a^b \frac{d }{dx}\left( \epsilon(x)\frac{\partial f}{\partial y'} \right) =0dx \nonumber\\
&= \epsilon(x)\frac{\partial f}{\partial y'} |_{x=a}^{x=b} + \int_a^b \epsilon(x)\left(  \frac{\partial f}{\partial y} - \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right) \right) dx  = 0\nonumber \\
&\text{recall that variation at the endpoints must be zero, thus the first term vanishes}\nonumber\\
\frac{dS}{d\alpha}|_{\alpha =0} &= \int_a^b \epsilon(x)\left(  \frac{\partial f}{\partial y} - \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right) \right) dx  = 0
\end{align}
$$
### assuming non-trivial $\epsilon(x) \neq 0, \forall x$, the integral is of the form $\int e(x) g(x)dx = 0$, for any arbitrary choice of $e(x)$, $g(x)$ must be 0 for all $x$. We can prove this simply by considering functions $e(x)$ and $g(x)$ whose signs are the same for some interval, this nessessarily means $\int e(x) g(x)dx \geq 0$, thus necessitating $g(x) = 0, \forall x \in (x_1, x_2)$. Thus:
$$
\frac{\partial f}{\partial y} - \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right) = 0
$$
### minimizes the action.


# **Example**
----
### Shortest Path between two points: $(x_1,y_1), (x_2,y_2)$, connected by $y(x)$.
Consider an infinitesimal line element along $y(x)$ given by $ds$. By the Pythagorean theorem, we know $(ds)^2 = (dx)^2+(dy)^2$, which yields $ds = \sqrt{(dx)^2+(dy)^2}$. In this case, $y$ is a function of $x$, so $x$ is the independent variable. Using this knowledge, we should expect to integrate the infinitesimals along the $x$-axis. To do this, we express $dy$ in terms of $dx$:
Since $y=y(x)$, the differential $dy$ is given by:
$$
\begin{align}
    dy = \frac{dy}{dx} dx
\end{align}
$$
This yields the following useful simplification for $ds$:
$$
\begin{align}
    ds &= \sqrt{(dx)^2+(dy)^2} \nonumber \\
       &= \sqrt{(dx)^2+\left(\frac{dy}{dx} dx\right)^2} \\
       &= \sqrt{(dx)^2 \left(1 +\left(\frac{dy}{dx}\right)^2\right)} \\
       &= \sqrt{ 1 +\left(\frac{dy}{dx}\right)^2} dx
\end{align}
$$
Thus, the total length of $y(x)$ from point 1 to point 2 is given by $L = \int ds$:
$$
\begin{align}
    L &= \int_1^2 ds \nonumber \\
       &= \int_{x_1}^{x_2} \sqrt{ 1 +\left(\frac{dy}{dx}\right)^2} dx
\end{align}
$$
The goal is to find the function $y(x)$ such that the length, $L$, is minimized. Using the Euler-Lagrange equation, the solution $y(x)$ must satisfy:
$$
\frac{\partial f}{\partial y} - \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right) = 0
$$
In this problem, we define the integrand (often called the Lagrangian or functional in this context, though here it's just the function whose integral we extremize) as $f(y, y', x)$. Letting $y' = \frac{dy}{dx}$:
$$
f(y, y', x) = \sqrt{ 1 +(y')^2}
$$
We start by calculating each term in the Euler-Lagrange equation:
$$
\begin{align}
\frac{\partial f}{\partial y} &= \frac{\partial }{\partial y}  \left(  \sqrt{ 1 +(y')^2} \right)\nonumber\\
\end{align}
$$
This partial derivative is zero because the functional $f$ requires $x,y,y'$ to be formally independent variables. When calculating $\frac{\partial f}{\partial y}$, $y'$ and $x$ are treated as independent variables held constant:
$$
\frac{\partial f}{\partial y} = 0
$$
For the second term:
$$
\begin{align}
\frac{\partial f}{\partial y'} &= \frac{\partial }{\partial y'} \left(  \sqrt{ 1 +\left(y'\right)^2} \right) \\
&= \frac{1}{2\sqrt{1+(y')^2}} \cdot (2y') \\
&= \frac{y'}{\sqrt{1+(y')^2}}
\end{align}
$$
Let's call this result $h(y') = \frac{y'}{\sqrt{1+(y')^2}}$.
Substituting these results back into the Euler-Lagrange equation:
$$
\begin{align}
\frac{\partial f}{\partial y} - \frac{d }{dx} \left( \frac{\partial f}{\partial y'}  \right) &= 0\\
(0) - \frac{d }{dx} \left( \frac{y'}{\sqrt{1+(y')^2}} \right) &= 0\\
\frac{d }{dx} \left( \frac{y'}{\sqrt{1+(y')^2}} \right) &= 0
\end{align}
$$
This implies that the term inside the derivative must be a constant:
$$
\frac{y'}{\sqrt{1+(y')^2}} = C
$$
where $C$ is an arbitrary constant. Now we can solve for $y'(x)$ and then $y(x)$:
Squaring both sides:
$$
\begin{align}
\frac{(y')^2}{1+(y')^2} &= C^2 \\
(y')^2 &= C^2 (1+(y')^2) \\
(y')^2 &= C^2 + C^2 (y')^2 \\
(y')^2 - C^2 (y')^2 &= C^2 \\
(y')^2 (1-C^2) &= C^2 \\
(y')^2 &= \frac{C^2}{1-C^2}
\end{align}
$$
Let $m^2 = \frac{C^2}{1-C^2}$. Since $C$ is an arbitrary constant, $m^2$ is also some constant (for $m$ to be real, $1-C^2$ must be positive, implying $C^2 < 1$. If $C^2=1$, the denominator is zero, corresponding to an infinite slope, i.e., a vertical line. If $C=0$, then $y'=0$).
So, $(y')^2 = m^2$, which means:
$$
y' = \pm m
$$
Since $m$ can absorb the sign, we can write $y' = m_{const}$, where $m_{const}$ is a new constant.
$$
\begin{align}
\frac{dy}{dx} &= m_{const} \\
\int dy &= \int m_{const} dx \\
y(x) &= m_{const}x + b
\end{align}
$$
where $b$ is another constant of integration.
We have therefore derived the familiar result that the function $y(x)$ minimizing the distance between two points in a plane (assuming a Euclidean metric) is a straight line.
